# **RESULTADOS OBTENIDOS**
En el presente notebook realizaremos una comparación de los resultados obtenidos por cada modelo en los 3 datasets creados (train, dev y test). 

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
```

```python
train = pd.read_csv("Dataset train.csv")
train.head()
```

```python
test = pd.read_csv("Dataset test.csv")
test.drop("TARGET", axis = 1, inplace = True)
test.head()
```

```python
cols = train.select_dtypes(exclude = ['object']).columns
cols
```

Obtenemos el rango de importancia de cada variable para utilizar las 50 primeras en el modelo de regresión logística.

```python
corr_df = train.corr()
corr_df.fillna(0, inplace = True)
corr_target = np.abs(corr_df['TARGET']).sort_values(ascending = False)
importance = corr_target.index.to_numpy()
importance = np.delete(importance, 0)

importance
```

Realizaremos la misma división del dataset train que en los anteriores notebooks:
* Train set: datos para entrenar los modelos (96%)
* Dev set: datos para comparar los modelos (2%: 11K)
* Test set: datos para probar los modelos (2%: 11K)

```python
pos_records = train.loc[train['TARGET'] == 1, cols]
pos_records.shape
```

```python
neg_records = train.loc[train['TARGET'] == 0, cols]
neg_records.shape
```

```python
X_pos = pos_records.loc[:, cols != "TARGET"]
Y_pos = pos_records.loc[:, "TARGET"]
X_neg = neg_records.loc[:, cols != "TARGET"]
Y_neg = neg_records.loc[:, "TARGET"]
```

```python
X_train_pos, X_rem_pos, Y_train_pos, Y_rem_pos = train_test_split(X_pos, Y_pos, test_size = 0.04, shuffle = True, random_state = 0)
X_dev_pos, X_test_pos, Y_dev_pos, Y_test_pos = train_test_split(X_rem_pos, Y_rem_pos, test_size = 0.5, shuffle = True, random_state = 0)

X_train_neg, X_rem_neg, Y_train_neg, Y_rem_neg = train_test_split(X_neg, Y_neg, test_size = 0.04, shuffle = True, random_state = 0)
X_dev_neg, X_test_neg, Y_dev_neg, Y_test_neg = train_test_split(X_rem_neg, Y_rem_neg, test_size = 0.5, shuffle = True, random_state = 0)

X_train = pd.concat([X_train_pos, X_train_neg], axis = 0)
Y_train = pd.concat([Y_train_pos, Y_train_neg], axis = 0)
X_dev = pd.concat([X_dev_pos, X_dev_neg], axis = 0)
Y_dev = pd.concat([Y_dev_pos, Y_dev_neg], axis = 0)
X_test = pd.concat([X_test_pos, X_test_neg], axis = 0)
Y_test = pd.concat([Y_test_pos, Y_test_neg], axis = 0)
```

```python
print("X_train size: {}".format(X_train.shape))
print("Y_train size: {}".format(Y_train.shape))
print("X_dev size: {}".format(X_dev.shape))
print("Y_dev size: {}".format(Y_dev.shape))
print("X_test size: {}".format(X_test.shape))
print("Y_test size: {}".format(Y_test.shape))
```

```python
print("Proportion train set: {:.3f} %".format(Y_train.value_counts()[1]/Y_train.shape[0] * 100))
print("Proportion dev set: {:.3f} %".format(Y_dev.value_counts()[1]/Y_dev.shape[0] * 100))
print("Proportion test set: {:.3f} %".format(Y_test.value_counts()[1]/Y_test.shape[0] * 100))
```

**IMPORTACIÓN DE LOS MODELOS PREENTRENADOS**

Ahora cargaremos los modelos obtenidos en los anteriores notebooks utilizando la librería *joblib*.

```python
import joblib

logistic_regresion = joblib.load('logistic_regresion_model.pkl')
random_forest = joblib.load('random_forest_model.pkl')
xgboost = joblib.load('xgboost_model.pkl')
```

```python
dict = {'Train' : (X_train, Y_train), 'Dev' : (X_dev, Y_dev), 'Test' : (X_test, Y_test)}
results = []

for name, sets in dict.items():
    X, Y = sets
    
    lr_score = roc_auc_score(Y, logistic_regresion.predict_proba(X[importance[:50]])[:, 1])
    rf_score = roc_auc_score(Y, random_forest.predict_proba(X)[:, 1])
    xgb_score = roc_auc_score(Y, xgboost.predict_proba(X)[:, 1])

    results.append([name, lr_score, rf_score, xgb_score])

results = np.array(results)
```

```python
results_df = pd.DataFrame(results, columns = ['Set', 'Regresión Logística', 'Random Forest', 'XGBoost'])

results_df['Regresión Logística'].astype(np.float32)
results_df['Random Forest'].astype(np.float32)
results_df['XGBoost'].astype(np.float32)

results_df.set_index('Set')
```

**CONCLUSIONES**

De los 3 modelos desarrollados, el modelo de random forest es el que presenta mayor score en los 3 sets, por lo tanto, será el modelo utilizado como solución de este proyecto.
